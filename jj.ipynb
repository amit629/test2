```python
import numpy as np

class FeedForwardNN:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        self.W_hidden = np.random.rand(hidden_size, input_size)
        self.b_hidden = np.zeros((hidden_size, 1))
        self.W_output = np.random.rand(output_size, hidden_size)
        self.b_output = np.zeros((output_size, 1))
        
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def sigmoid_derivative(self, x):
        return x * (1 - x)
    
    def forward_propagation(self, x):
        z_hidden = np.dot(self.W_hidden, x) + self.b_hidden
        a_hidden = self.sigmoid(z_hidden)
        
        z_output = np.dot(self.W_output, a_hidden) + self.b_output
        a_output = self.sigmoid(z_output)
        
        return a_hidden, a_output
    
    def backpropagation(self, x, y, learning_rate):
        a_hidden, a_output = self.forward_propagation(x)
        y = y.reshape(a_output.shape)
        
        error_output = y - a_output
        delta_output = error_output * self.sigmoid_derivative(a_output)
        
        error_hidden = np.dot(self.W_output.T, delta_output)
        delta_hidden = error_hidden * self.sigmoid_derivative(a_hidden)
        
        self.W_output += learning_rate * np.dot(delta_output, a_hidden.T)
        self.b_output += learning_rate * delta_output
        
        self.W_hidden += learning_rate * np.dot(delta_hidden, x.T)
        self.b_hidden += learning_rate * delta_hidden

input_size = 3
hidden_size = 4
output_size = 2

nn = FeedForwardNN(input_size, hidden_size, output_size)

x_example = np.array([0.1, 0.2, 0.3])
y_example = np.array([0.4, 0.5])

learning_rate = 0.1
nn.backpropagation(x_example, y_example, learning_rate)
```
